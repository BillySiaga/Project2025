```python
#Selecting strategy to handle missing values, and tell us why you had made that choice.
# replacing missing values with ffill and bfill method because of the repetative values of the column
#
df['reading_remarque'] = df['reading_remarque'].fillna(method = 'ffill').fillna(method = 'bfill', inplace =True)
print(df['reading_remarque'].value_counts())
```

## Transform the 'counter_type' feature to a numeric variable using the encoder of your choice.

```python
#Transform the 'counter_type' feature to a numeric variable using the encoder of your choice.
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder() # Initialize the encoder

# Apply the transformation
df['counter_type_numeric'] = label_encoder.fit_transform(df['counter_type'])

print(df[['counter_type', 'counter_type_numeric']].head())
```
