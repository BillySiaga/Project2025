```python
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
te=TransactionEncoder()
te_ary=te.fit(toy_dataset).transform(toy_dataset)    #Apply one-hot-encoding on our dataset
toy_dataset=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
toy_dataset
```

```python
#explore the checkpoint dataset using Pandas and Plotly.
import plotly.express as px
fig = px.treemap(toy_dataset, path=['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'])
fig.show()
```

```python
# selecting items with a minimum of 60% support
from mlxtend.frequent_patterns import apriori
apriori(toy_dataset, min_support=0.6)
```

## Interpretation of the Results
From the apriori algorithm results, 5 and 7 (pants and scarfs) respectively were the most frequent with a support of 0.8 and occur in 60% of transactions. this suggest a strong correlation that could be used in recommendation system and market basket analysis.
The business can thus consider generating an asosication rule where if item 5 is present then is likely that item 7 will be present. evaluating the confidence and lift of this rule would be important to gauge the strength of the associations.
