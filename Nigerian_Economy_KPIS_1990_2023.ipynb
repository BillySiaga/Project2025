# 4. Rename our datasets columns

```python
print(list(df.columns))
```

```python
df.rename(columns = {'Unemployment': 'Unemployment Rate (%)',
                     'Government debt': 'Government Debt Rate (%)',
                     'Agriculture': 'Agriculture GDP (Naira)',
                     'Industry': 'Industry GDP (Naira)',
                     'Services': 'Services GDP (Naira)',
                     'Leadership': 'Presidents',
                     'Inflation rate': 'Inflation Rate (%)'
                     },inplace = True)
df.head()
```

# 5. Perform tranformation on the columns

```python
#lets transform the inflation rate, unemployment rate and Governmentdebt rate into percentages
df['Inflation Rate (%)'] =df['Inflation Rate (%)'].apply(lambda x: x*100)
df['Unemployment Rate (%)'] = df['Unemployment Rate (%)'].apply(lambda x: x*100)
df['Government Debt Rate (%)']= df['Government Debt Rate (%)'].apply(lambda x: x*100)
#
df.head()
```

# 6. Apply Basic Filter

```python
#unique presidents
unique_president = df['Presidents'].unique()
print(unique_president)
```

```python
df.loc[df['Presidents'] == 'Goodluck Ebele Jonathan']
```

```python
df.loc[df['Presidents'] == 'Sani Abacha']
```

```python
df.loc[df['Presidents'] == 'Buhari']
```

```python
for president in unique_president:
  result = df.loc[df['Presidents'] == president]
  print(f'\nRecords for President: {president}')
  print(result.to_string(index =False))
```

```python
highest_inflation_year = df.loc[df['Inflation Rate (%)'] ==df['Inflation Rate (%)'].max(), 'Year'].values[0]
print(f"The year with the highest inflation rate is: {highest_inflation_year}")
highest_inflation_by_president = df.groupby('Presidents')['Inflation Rate (%)'].max()
#
print("\nHighest Inflation rate for each president")
print(highest_inflation_by_president)
```

```python
#lowest Inflation year
lowest_inflation_year = df.loc[df['Inflation Rate (%)'] ==df['Inflation Rate (%)'].min(), 'Year'].values[0]
print(f"The year with the lowest inflation rate is: {highest_inflation_year}")
lowest_inflation_by_president = df.groupby('Presidents')['Inflation Rate (%)'].min()
#
print("\nHighest Inflation rate for each president")
print(lowest_inflation_by_president)
```

```python
#grouping by specific values
#calculate the unemployment rate for each Leadership era
df.groupby('Presidents')[['Inflation Rate (%)', 'Unemployment Rate (%)', 'Government Debt Rate (%)']].mean()
```

# Visualization of the Insights

```python
#lets cretae a simple visualization to see unemployment rate over the years in Nigeria
#label the x and y axis
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.title("Unemployment Rate over Years")

df.groupby("Year")["Unemployment Rate (%)"]. mean().plot();
```

```python
#lets cretae a simple visualization to see average unemployment rate by President
#label the x and y axis
plt.xlabel("Presidents")
plt.ylabel("Unemployment Rate (%)")
plt.title(" Average Unemployment Rate by President")

df.groupby("Presidents")["Unemployment Rate (%)"]. mean().plot(kind = "barh");
```

```python
# comparing inflation rate, unemployment rate and government debt
df.groupby("Year")[["Inflation Rate (%)", "Unemployment Rate (%)", "Government Debt Rate (%)"]].mean().plot()
plt.xlabel("Years")
plt.ylabel("Average Rate (%)")
plt.title("Economy KPIs Over Years");
```

```python
#create inflation rate percentge change
df["Inflation Rate Change"] = df["Inflation Rate (%)"].pct_change()
#plot the change
df.plot(x= "Year", y = "Inflation Rate Change", kind = "bar");
```
